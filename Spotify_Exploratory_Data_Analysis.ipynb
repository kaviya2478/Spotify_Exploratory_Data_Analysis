```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks = pd.read_csv('tracks.csv')
df_artists = pd.read_csv('artists.csv')
df_combined = pd.concat([df_tracks, df_artists], ignore_index=True)
df_combined.head()
```

```python
pd.isnull(df_combined).sum()
```

```python
df_combined.info()
```

**Top 10 Least Popular Song**

```python
sorted_df = df_combined.sort_values('popularity', ascending=True).head(10)
sorted_df
```

**Statistics Description**

```python
df_combined.describe().transpose()
```

**Top 10 songs popularity more than 90**

```python
most_popular=df_combined.query('popularity>90', inplace=False).sort_values('popularity',ascending=False)
most_popular[:10]
```

```python
df_combined[['artists']].iloc[18]
```

**Correlation**

```python
sample_df=df_combined.sample(int(0.004*len(df_combined)))
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y='loudness',x='energy',color='c').set(title='Loudness vs Energy Correlation')
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y='popularity',x='acousticness',color='m').set(title='Popularity vs Acousticness Correlation')
```

```python
df_genre=pd.read_csv('SpotifyFeatures.csv')
df_genre.head()
```

```python
plt.title("Duration of the Songs in Different Genres")
sns.color_palette("rocket", as_cmap=True)
sns.barplot(y='genre',x='duration_ms',data=df_genre)
plt.xlabel("Duration in ms")
plt.ylabel("Genres")
```

```python
sns.set_style(style='darkgrid')
plt.figure(figsize=(10,5))
famous=df_genre.sort_values('popularity',ascending=False)[:10]
sns.barplot(y='genre',x='popularity',data=famous).set(title='Top 5 Genres  by Popularity')
plt.show()
```
